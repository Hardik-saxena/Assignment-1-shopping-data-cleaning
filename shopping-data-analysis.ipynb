```python
df.fillna("Unknown", inplace=True)
```

```python
df = df.drop_duplicates()

print("Duplicates removed")
```

```python
print(df.columns)
```

```python
df[['title', 'seller_name', 'final_price']]
```

```python
# Convert final_price to numeric
df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')

# Convert initial_price also
df['initial_price'] = pd.to_numeric(df['initial_price'], errors='coerce')
```

```python
filtered_df = df[df['final_price'] > 1000]

filtered_df.head()
```

```python
df['discount_amount'] = df['initial_price'] - df['final_price']

df[['initial_price', 'final_price', 'discount_amount']].head()
```

```python
df.to_csv('/kaggle/working/cleaned_shopping_data.csv', index=False)

print("Cleaned dataset saved successfully!")
```
